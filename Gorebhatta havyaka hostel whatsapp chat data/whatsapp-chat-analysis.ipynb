```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with GBH Hostel AHMS.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern=r'\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}\s[ap]m'
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
print(dates)
```

```python
cleaned_dates=[entry.replace('\u202f','')for entry in dates]
cleaned_dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':cleaned_dates})

df['message_date'] = pd.to_datetime(df['message_date'],format='%d/%m/%y, %I:%M%p'
)
df.rename(columns={'message-date':'date'},inplace=True)

df.head()
```

```python
df.shape
```

```python
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user']=users
df['messages']=messages
df.drop(columns=['user_message'],inplace=True)

df.head(20)
```

```python
df['year']=df['message_date'].dt.year
df.head()
```

```python
df['month']=df['message_date'].dt.month_name()
df.head()
```

```python
df['day_of_month']=df['message_date'].dt.day
df.head()
```

```python
df['hour']=df['message_date'].dt.hour
df.head()
```

```python
df['minute']=df['message_date'].dt.minute
df.head()
```
